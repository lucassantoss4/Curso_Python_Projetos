```python
# Define uma classe chamada "Pessoa"
class Pessoa:
    def __init__(self, nome):
        self.nome = nome
        self.acordado = False
        self.comendo = False
        self.dirigindo = False

    def acordar(self):
        if self.acordado:
            print('A pessoa já está acordada')
        else:
            self.acordado = True
            print('A pessoa acordou')

    def comer(self):
        if self.comendo:
            print('A pessoa já está comendo')
        else:
            self.comendo = True
            print('A pessoa está comendo')

    def dirigir(self):
        if self.dirigindo:
            print('A pessoa já está dirigindo')
        else:
            self.dirigindo = True
            print('A pessoa está dirigindo')

# Cria uma instância da classe Pessoa
joao = Pessoa('João')

# Chama os métodos corretos
joao.acordar()  # A pessoa acordou
joao.acordar()  # A pessoa já está acordada

joao.comer()     # A pessoa está comendo
joao.comer()     # A pessoa já está comendo

joao.dirigir()   # A pessoa está dirigindo
joao.dirigir()   # A pessoa já está dirigindo
```

```python
class Evento:
    def __init__(self, capacidade_inicial=3):
        self.capacidade_inicial = capacidade_inicial
        self.lugares_ocupados = 0
        self.lugares_disponiveis = self.capacidade_inicial

    def cancelar(self):
        if self.lugares_ocupados > 0:
            self.lugares_ocupados -= 1
            self.lugares_disponiveis += 1
            return "Reserva cancelada com sucesso!"
        else:
            return "Não existem reservas para cancelar!"

    def reservar(self):
        if self.lugares_disponiveis > 0:
            self.lugares_ocupados += 1
            self.lugares_disponiveis -= 1
            return "Lugar reservado com sucesso!"
        else:
            return "Não existem mais lugares disponíveis para reserva!"

    def exibir_lugares_disponiveis(self):
        return f"Lugares disponíveis: {self.lugares_disponiveis}"


# Criando a instância do evento
evento = Evento()

# Loop de interação com o usuário
while True:
    opcao = int(input("Selecione uma Opção:\n1 - Exibir lugares\n2 - Reservar Lugar\n3 - Cancelar Reserva\n0 - Sair\n"))


    # lugar_diponivel = evento.exibir_lugares_disponiveis()

    if evento.lugares_disponiveis == 0:
        print("Não existem mais lugares disponíveis para reserva. Encerrando o programa.")
        break

    if opcao == 0 :
        break
    elif opcao == 1:
        print(evento.exibir_lugares_disponiveis())
    elif opcao == 2:
        print(evento.reservar())
    elif opcao == 3:
        print(evento.cancelar())
    else:
        print("Opção inválida. Por favor, selecione uma opção válida.")
```

```python
class ManipuladorDeListas:
    def __init__(self):
        self.lista = list()
        
    def adicionar_elemento(self, elemento):  # Adicionei 'self' como primeiro parâmetro
        self.lista.append(elemento)  # Corrigido de 'apped' para 'append'
        return "Elemento adicionado com sucesso!"

    def remover_elemento(self,elemento):
        self.lista.remove(elemento)
        return 'Elemento removido com sucesso!'

    def maior_elemento(self):
        maior = max(self.lista)
        return f'O maior elemento da lista é {maior}'

    def manor_elemento(self):
        menor = min(self.lista)
        return f'O manor elemento da lista é {menor}'

    def media_lista(self):
        if self.lista:
            soma = sum(self.lista)
            media = soma / len(self.lista)
            return f"A média da soma dos elementos da lista é {media}"
        else:
            return "A lista está vazia."


    def mostra_lista(self):
        lista_amostra = self.lista
        return f'Lista {lista_amostra}'



# Criando uma instância de ManipuladorDeListas
Manipular = ManipuladorDeListas()

# Loop de interação com o usuário
while True:
    opcao = int(input("Selecione uma Opção:\n1 - Adicionar Elementos\n2 - Remover Elementos\n3 - Encontrar Maior Elemento \n4 - Encontrar Maior Elemento \n5 - calcular média \n6 - Mostra Lista \n0 - Sair\n"))

    if opcao == 0 :
        break
    elif opcao == 1:    
        try:
            num = input(int('Digite um Número: '))
            print(Manipular.adicionar_elemento(num))
        except ValueError:
            print("Entrada inválida. Por favor, digite um número. ")
            continue
    elif opcao == 2:
            try:
                num = input(int('Digite um Número: '))
                print(Manipular.remover_elemento(num))
            except ValueError:
                print("Entrada inválida. Por favor, digite um número. ")
            continue
    elif opcao == 3:
        print(Manipular.maior_elemento())

    elif opcao == 4:
        print(Manipular.manor_elemento())
    
    elif opcao == 5:
        print(Manipular.media_lista())

    elif opcao == 6:
        print(Manipular.mostra_lista())
    else:
        print("Opção inválida. Por favor, selecione uma opção válida.")
```
