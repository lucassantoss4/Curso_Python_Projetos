```python
class ManipuladorDeListas:
    def __init__(self):
        self.lista = list()
        
    def adicionar_elemento(self, elemento):
        self.lista.append(elemento)
        return "Elemento adicionado com sucesso!"

    def remover_elemento(self, elemento):
        if elemento in self.lista:
            self.lista.remove(elemento)
            return "Elemento removido com sucesso!"
        else:
            return "Elemento não encontrado na lista."

    def maior_elemento(self):
        if self.lista:
            maior = max(self.lista)
            return f"O maior elemento da lista é {maior}"
        else:
            return "A lista está vazia."

    def menor_elemento(self):  # Corrigido de 'manor_elemento' para 'menor_elemento'
        if self.lista:
            menor = min(self.lista)
            return f"O menor elemento da lista é {menor}"
        else:
            return "A lista está vazia."

    def media_lista(self):
        if self.lista:
            soma = sum(self.lista)
            media = soma / len(self.lista)
            return f"A média dos elementos da lista é {media}"
        else:
            return "A lista está vazia."

    def mostra_lista(self):
        return f"Lista: {self.lista}"


# Criando uma instância de ManipuladorDeListas
Manipular = ManipuladorDeListas()

# Loop de interação com o usuário
while True:
    try:
        opcao = int(input("Selecione uma Opção:\n1 - Adicionar Elementos\n2 - Remover Elementos\n3 - Encontrar Maior Elemento\n4 - Encontrar Menor Elemento\n5 - Calcular Média\n6 - Mostrar Lista\n0 - Sair\n"))
    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")
        continue

    if opcao == 0:
        break
    elif opcao == 1:
        try:
            num = int(input("Digite um número: "))
            print(Manipular.adicionar_elemento(num))
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")
    elif opcao == 2:
        try:
            num = int(input("Digite um número: "))
            print(Manipular.remover_elemento(num))
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")
    elif opcao == 3:
        print(Manipular.maior_elemento())
    elif opcao == 4:
        print(Manipular.menor_elemento())
    elif opcao == 5:
        print(Manipular.media_lista())
    elif opcao == 6:
        print(Manipular.mostra_lista())
    else:
        print("Opção inválida. Por favor, selecione uma opção válida.")
```
